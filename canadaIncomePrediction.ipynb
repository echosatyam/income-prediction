```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model as lm
```

```python
df= pd.read_csv('data/canada_per_capita_income.csv')
df
```

```python
plt.plot(df['year'],df['income'],'b.')
plt.xlabel('Year')
plt.ylabel('Income')
plt.show()
```

```python
reg = lm.LinearRegression()
reg.fit(df[['year']],df.income)
```

```python
reg.predict([[2020]])
```

```python

plt.xlabel('Year')
plt.ylabel('Income')
plt.scatter(df['year'],df['income'],marker=".",color='green')
plt.plot(df['year'],reg.predict(df[['year']]),'r-')
plt.show()
```

```python
canadaDF= pd.DataFrame({'Year':range(2000,2021,1)})
```

```python
canadaDF['Income'] = reg.predict(canadaDF[['Year']])
```

```python
canadaDF.to_csv('data/prediction.csv',sep=' ')
```
